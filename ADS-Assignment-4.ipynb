```python
bouquet_packages = ['Basic','Plus','Premium']

df['bouquetpackages'] = pd.qcut(df['monthlycharges'],q=3, precision=0, labels=bouquet_packages)

np.round(pd.pivot_table(df[df['churn'] == "No"], 
               margins = True, 
               margins_name = "Total", 
               index=['churn'], 
               columns = 'bouquetpackages', 
               values = 'monthlycharges', 
               aggfunc=np.mean, 
               fill_value = 0))
```
